```python
import matplotlib.pyplot as plt

# create two sets of points
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 4, 5, 6]

x2 = [1, 2, 3, 4, 5]
y2 = [6, 5, 4, 3, 2]

# plot the points
plt.scatter(x1, y1, color='red', label='Set 1')
plt.scatter(x2, y2, color='blue', label='Set 2')

# draw lines between the corresponding points
for i in range(len(x1)):
    plt.plot([x1[i], x2[i]], [y1[i], y2[i]], color='gray')

# set the title and legend
plt.title('Two Sets of Points with Lines')
plt.legend()

# show the plot
plt.show()
```
